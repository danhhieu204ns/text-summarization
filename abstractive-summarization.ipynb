```python
from transformers import pipeline
from datasets import load_dataset
```

```python

dataset = load_dataset("OpenHust/vietnamese-summarization")
```

```python
from datasets import DatasetDict
```

```python
split_ds = dataset["train"].train_test_split(test_size = 0.2,seed = 42)
train_ds = split_ds["train"]
valid_ds = split_ds["test"]
split_valid = valid_ds.train_test_split(test_size = 0.5,seed =42)
valid_ds = split_valid["train"]
test_ds = split_valid["test"]
ds = DatasetDict({
  "train": train_ds,
    "validation": valid_ds,
    "test": test_ds
})
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

model_name = "VietAI/vit5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)
```

```python
def convert_examples_to_features(example_batch):
    inputs = ["summarize: " + doc for doc in example_batch["Document"]]
    input_encodings = tokenizer(inputs,max_length = 1024,truncation = True)
    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch["Summary"],max_length = 128,truncation = True)
    return {
        "input_ids":input_encodings["input_ids"],
        "attention_mask":input_encodings["attention_mask"],
        "labels":target_encodings["input_ids"]
    }
```

```python
from transformers import DataCollatorForSeq2Seq
seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
from transformers import TrainingArguments, Trainer
from transformers import Seq2SeqTrainingArguments
import evaluate
```

```python
import numpy as np
```

```python
from peft import LoraConfig, get_peft_model
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)
peft_cfg = LoraConfig(
    r=32,                    # rank cao hơn -> học được nhiều đặc trưng hơn
    lora_alpha=64,           # scale tương ứng
    lora_dropout=0.05,       # regularization nhẹ
    bias="none",             # không fine-tune bias
    task_type="SEQ_2_SEQ_LM",# bắt buộc cho ViT5 / T5
    # target_modules=["q","k","v","o"]
    target_modules=["q","k","v","o","wi_0","wi_1","wo"]

)
model = get_peft_model(model, peft_cfg)
model.print_trainable_parameters()
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="t5_new_lora",
    num_train_epochs=1,
    learning_rate=5e-5,
    weight_decay=0.01,
    warmup_steps=500,
    lr_scheduler_type="linear",
    logging_steps=50,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=8,
    fp16=True,

    # 🚫 Tắt eval hoàn toàn:
    eval_strategy="no",      # hoặc eval_strategy="no" nếu bạn dùng tên rút gọn
    save_strategy="epoch",         # chỉ lưu cuối mỗi epoch, không kèm eval
    load_best_model_at_end=False,  # tắt vì không có đánh giá để so sánh

    optim="adafactor",
    dataloader_num_workers=8,
    dataloader_pin_memory=True,
    dataloader_persistent_workers=True,
    metric_for_best_model="rougeL",
    greater_is_better=True,
    predict_with_generate=False,
    report_to="none",
    push_to_hub=False,
    group_by_length=True,
    remove_unused_columns=False,
    label_smoothing_factor=0.1,
)
```

```python
dataset_tp = ds.map(convert_examples_to_features,batched = True)
columns = ["input_ids","labels","attention_mask"]
dataset_tp.set_format(type="torch",columns = columns)
```

```python
def build_compute_metrics(tokenizer):
    pad_id = tokenizer.pad_token_id
    def compute_metrics(eval_pred):
        preds, labels = eval_pred
        if isinstance(preds, tuple):
            preds = preds[0]
        decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True, clean_up_tokenization_spaces=True)
        labels = np.where(labels != -100, labels, pad_id)
        decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True, clean_up_tokenization_spaces=True)
        decoded_preds  = [p.strip() for p in decoded_preds]
        decoded_labels = [l.strip() for l in decoded_labels]
        result = rouge_metric.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)
        # Trả về các khóa 'rouge1','rouge2','rougeL','rougeLsum' (Trainer sẽ log kèm tiền tố 'eval_')
        return {k: result[k] for k in ["rouge1", "rouge2", "rougeL", "rougeLsum"]}
    return compute_metrics

compute_metrics = build_compute_metrics(tokenizer)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    tokenizer=tokenizer,
    data_collator=seq2seq_data_collator,
    train_dataset=dataset_tp["train"],
    eval_dataset=dataset_tp["validation"],
    compute_metrics=compute_metrics

)
```

```python
from transformers.integrations import WandbCallback
trainer.callback_handler.callbacks = [
    cb for cb in trainer.callback_handler.callbacks if not isinstance(cb, WandbCallback)
]
import os
os.environ["WANDB_DISABLED"] = "true"
```

```python
trainer.train()
```

```python
# !pip install evaluate sacrebleu  # nếu chưa cài
import torch
from torch.utils.data import DataLoader
from transformers import DataCollatorForSeq2Seq
import evaluate
from tqdm.auto import tqdm

# 1) Chuẩn bị dataset test đã map
test_dataset = dataset_tp["test"]     # nếu tên split khác (e.g. "validation"), đổi ở đây

# 2) Collator để pad và tạo batch đúng cách
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model, padding="longest")

# 3) DataLoader
BATCH_SIZE = 8  # giảm nếu GPU yếu (2–4), tăng nếu đủ VRAM
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, collate_fn=data_collator)

# 4) Metric
rouge = evaluate.load("rouge")
sacrebleu = evaluate.load("sacrebleu")

# 5) Sinh dự đoán
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

all_preds, all_refs = [], []

# Lấy reference gốc từ ds (tránh decode từ labels có -100)
raw_refs = ds["test"]["Summary"]

i_ref = 0
with torch.no_grad():
    for batch in tqdm(test_loader, desc="Generating"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        # Tinh chỉnh tham số generate tùy mô hình/dung lượng
        gen_out = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_new_tokens=128,            # cùng giới hạn như khi preprocess
            num_beams=4,                   # beam search ổn định cho đánh giá
            length_penalty=1.0,
            no_repeat_ngram_size=3,
            early_stopping=True
        )

        preds = tokenizer.batch_decode(gen_out, skip_special_tokens=True)
        # match số lượng ref tương ứng batch hiện tại
        refs = raw_refs[i_ref:i_ref+len(preds)]
        i_ref += len(preds)

        # làm sạch nhẹ
        preds = [p.strip() for p in preds]
        refs  = [r.strip() for r in refs]

        all_preds.extend(preds)
        all_refs.extend(refs)

# 6) Tính ROUGE (dùng stemmer cho Rouge1/2/L phổ biến)
rouge_results = rouge.compute(predictions=all_preds, references=all_refs, use_stemmer=True)

# 7) Tính BLEU (sacreBLEU yêu cầu references dạng list-of-lists)
bleu_results = sacrebleu.compute(predictions=all_preds, references=[[r] for r in all_refs])

print("ROUGE:")
for k, v in rouge_results.items():
    # nhân 100 để rõ ràng hơn
    print(f"  {k}: {v*100:.2f}")

print(f"\nBLEU (sacreBLEU): {bleu_results['score']:.2f}")
```
